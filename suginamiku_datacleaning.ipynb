```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns

%matplotlib inline

print('Libraries imported.')
```

```python
suginami_data = pd.read_csv("suginamiku_2010_2020.csv", encoding= 'unicode_escape')

suginami_data
```

```python
missing_values_count = suginami_data.isnull().sum()
missing_values_count
```

```python
del_list = ["No","Region","Transaction-price(Unit price m^2)","Purpose of Use",
           "Land shape","Frontage","Layout","Total floor area(m^2)",
           "Frontage roadFDirection","Frontage roadFClassification",
           "Frontage roadFBreadth(m)","City,Town,Ward,Village code",
           "Prefecture","City,Town,Ward,Village","Renovation","Transactional factors"]
```

```python
for i in del_list:
    del suginami_data[i]
```

```python
suginami_data
```

```python
suginami_data["Type"].value_counts()
```

```python
# keep only pre-owned condominiums

suginami_data.drop(suginami_data.loc[suginami_data["Type"] == "Residential Land(Land and Building)"].index, inplace = True)
```

```python
suginami_data.drop(suginami_data.loc[suginami_data["Type"] == "Residential Land(Land Only)"].index, inplace = True)

suginami_data["Type"].value_counts()
```

```python
del suginami_data["Use"]
```

```python
suginami_data.isna().sum()
```

```python
median_coverage_ratio = suginami_data["Maximus Building Coverage Ratio(%)"].median()
print(median_coverage_ratio)
suginami_data["Maximus Building Coverage Ratio(%)"] = suginami_data["Maximus Building Coverage Ratio(%)"].replace(np.nan, median_coverage_ratio)
#suginami_data["Maximus Building Coverage Ratio(%)"] = suginami_data["Maximus Building Coverage Ratio(%)"]/100
```

```python
median_floor_area_ratio = suginami_data["Maximus Floor-area Ratio(%)"].median()
print(median_floor_area_ratio)
suginami_data["Maximus Floor-area Ratio(%)"] = suginami_data["Maximus Floor-area Ratio(%)"].replace(np.nan, median_floor_area_ratio)
#suginami_data["Maximus Floor-area Ratio(%)"] = suginami_data["Maximus Floor-area Ratio(%)"]/100
```

```python
suginami_data.dropna(subset=["City Planning"], axis=0, inplace=True)
```

```python
suginami_data.drop(suginami_data.loc[suginami_data["Nearest stationFDistance(minute)"] == "2H-"].index, inplace = True)
```

```python
median_time = suginami_data["Nearest stationFDistance(minute)"].median()
print(median_time)
suginami_data["Nearest stationFDistance(minute)"] = suginami_data["Nearest stationFDistance(minute)"].replace(np.nan, median_time)
```

```python
suginami_data.dropna(subset=["Building structure"], axis=0, inplace=True)
```

```python
suginami_data = suginami_data.replace("before the war", np.nan)

median_year = suginami_data["Year of construction"].median()
print(median_year)
suginami_data["Year of construction"] = suginami_data["Year of construction"].replace(np.nan, median_year)
```

```python
suginami_data.isna().sum()
```

```python
suginami_data
```

```python
suginami_data.dtypes
```

```python
# change the data types

suginami_data[["Nearest stationFDistance(minute)", "Area(m^2)"]] = suginami_data[["Nearest stationFDistance(minute)", "Area(m^2)"]].astype("int")
suginami_data["Year of construction"] = suginami_data["Year of construction"].astype("int")
suginami_data["Transaction-price(total)"] = suginami_data["Transaction-price(total)"].astype("float")

suginami_data.dtypes
```

```python
# add "Building Age" column

suginami_data["Building Age"] = 2020 - suginami_data["Year of construction"]

suginami_data.head(2)
```

```python
column_names = list(suginami_data.columns.values.tolist())
column_names
```

```python
new_column_names = [ 'Type',
                     'Transaction period',
                     'Building structure',
                     'City Planning',
                     'Year of construction',
                     'Area',
                     'Nearest station\x81FName',
                     'Nearest station\x81FDistance(minute)',
                     'Area(m^2)',
                     'Building Age',
                     'Maximus Building Coverage Ratio(%)',
                     'Maximus Floor-area Ratio(%)',
                     'Transaction-price(total)']
suginami_data = suginami_data.reindex(columns=new_column_names)
suginami_data.head(2)
```

```python
# rename columns

suginami_data = suginami_data.rename(columns={"Type":"type","Area":"location",
                                            "Nearest stationFName":"nearest_station",
                                            "Nearest stationFDistance(minute)":"distance_to_station",
                                            "Area(m^2)":"area","Year of construction":"construction_year",
                                            "Building structure":"building_structure",
                                            "City Planning":"city_planning",
                                            "Maximus Building Coverage Ratio(%)":"maximum_coverage_ratio",
                                            "Maximus Floor-area Ratio(%)":"maximum_floor_area_ratio",
                                            "Transaction period":"transaction_period",
                                            "Building Age":"building_age","Transaction-price(total)":"price"})
suginami_data.head(1)
```

```python
suginami_data.to_csv("suginamiku_cleaned_data.csv")
```
