```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns

%matplotlib inline

print('Libraries imported.')
```

```python
suginami_data = pd.read_csv("suginamiku_cleaned_data.csv", index_col=0)
suginami_data.head(2)
```

```python
suginami_data.reset_index(drop=True, inplace=True)
suginami_data.head(2)
```

```python
# convert the price from JPY to USD
# conversion rate = 0.0097 (Jan 3th, 2021)

suginami_data["price"] = suginami_data["price"]*0.0097
suginami_data.head(2)
```

```python
suginami_data.drop(suginami_data.loc[suginami_data["price"]>1000000].index, inplace=True)

suginami_data
```

```python
suginami_data.reset_index(drop=True, inplace=True)
suginami_data.head(2)
```

```python
# data normalization

def data_norm(df):
    df = (df - df.min())/(df.max()-df.min())
    return df

attributes_list = ["distance_to_station","area","maximum_coverage_ratio","maximum_floor_area_ratio","building_age","price"]

data_norm = data_norm(suginami_data[attributes_list])

data_norm
```
