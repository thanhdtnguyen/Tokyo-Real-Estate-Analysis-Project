```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from utils import *
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, LambdaCallback

%matplotlib inline

print('Libraries imported.')
```

```python
toshima_data = pd.read_csv("toshimaku_2010_2020.csv", encoding= 'unicode_escape')

toshima_data.head(10)
```

```python
toshima_data
```

```python
missing_values_count = toshima_data.isnull().sum()
missing_values_count
```

```python
del toshima_data["Region"]
del toshima_data["Transaction-price(Unit price m^2)"]
del toshima_data["Purpose of Use"]
del toshima_data["Land shape"]
del toshima_data["Frontage"]
del toshima_data["Total floor area(m^2)"]
```

```python
del toshima_data["Frontage roadFDirection"]
del toshima_data["Frontage roadFClassification"]
del toshima_data["Frontage roadFBreadth(m)"]
```

```python
toshima_data
```

```python
toshima_data.isna().sum()
```

```python
median_coverage_ratio = toshima_data["Maximus Building Coverage Ratio(%)"].median()
print(median_coverage_ratio)
toshima_data["Maximus Building Coverage Ratio(%)"].replace(np.nan, median_coverage_ratio)
```

```python
median_floor_area_ratio = toshima_data["Maximus Floor-area Ratio(%)"].median()
print(median_floor_area_ratio)
toshima_data["Maximus Floor-area Ratio(%)"].replace(np.nan, median_floor_area_ratio)
```

```python
toshima_data.dropna(subset=["City Planning"], axis=0, inplace=True)
```

```python
toshima_data.dropna(subset=["Maximus Building Coverage Ratio(%)", "Maximus Floor-area Ratio(%)"], axis=0, inplace=True)
```

```python
toshima_data.dropna(subset=["Nearest stationFDistance(minute)"], axis=0, inplace=True)
```

```python
toshima_data.isna().sum()
```

```python
toshima_data = toshima_data.replace("before the war", np.nan)
```

```python
median_year = toshima_data["Year of construction"].median()
print(median_year)
toshima_data["Year of construction"] = toshima_data["Year of construction"].replace(np.nan, median_year)
```

```python
# delete unnecessary columns

del toshima_data["City,Town,Ward,Village code"]
del toshima_data["Prefecture"]
del toshima_data["Transactional factors"]

print("Columns Deleted")
```

```python
toshima_data.isna().sum()
```

```python
toshima_data["Layout"] = toshima_data["Layout"].replace(np.nan, "Unknown")
toshima_data["Renovation"] = toshima_data["Renovation"].replace(np.nan, "Unknown")
```

```python
toshima_data["Building structure"] = toshima_data["Building structure"].replace(np.nan, "Unknown")
toshima_data["Use"] = toshima_data["Use"].replace(np.nan, "Other")
```

```python
del toshima_data["No"]
toshima_data.isna().sum()
```

```python
toshima_data
```

```python
column_names = list(toshima_data.columns.values.tolist())
new_column_names = ['Type',
 'City,Town,Ward,Village',
 'Area',
 'Nearest station\x81FName',
 'Nearest station\x81FDistance(minute)',
 'Layout',
 'Area(m^2)',
 'Year of construction',
 'Building structure',
 'Use',
 'City Planning',
 'Maximus Building Coverage Ratio(%)',
 'Maximus Floor-area Ratio(%)',
 'Transaction period',
 'Renovation',
 'Transaction-price(total)']
toshima_data = toshima_data.reindex(columns=new_column_names)
toshima_data.head()
```

```python
toshima_data.to_csv("toshimaku_cleaned_data.csv")
```
